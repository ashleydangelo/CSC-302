Create a multipanel plot which draws density plots for each class-survived combination. Please also use the hue parameter based on the Sex attribute to have two density plots in each panel with different colors.

```python
# prompt: create multipanel plot which draws density plots for each class survived combination. Use hue parameter based on sex attribute to have density plots in each panel with different colors

df = pd.read_csv(titanic)
df['Survived'].replace({1:"Survived"}, inplace=True)
df = df[df["Pclass"]!='?']
g = sns.FacetGrid(df, col="Pclass", row="Survived", hue="Sex")
g.map(sns.kdeplot, "Age")
g.add_legend()
plt.show()
```

```python
# prompt: titles = [died in class 3, died in class 1, died in class 2, survived in class 3, survived in class 1, survived in class 2. for ax, title in zip(g.axes.flatten(), titles) ax.set_title(title)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')

titanic = ("/content/drive/MyDrive/DATA/titanic.csv")

df = pd.read_csv(titanic)
df['Survived'].replace({1:"Survived"}, inplace=True)
df = df[df["Pclass"]!='?']
g = sns.FacetGrid(df, col="Pclass", row="Survived", hue="Sex")
g.map(sns.kdeplot, "Age")
g.add_legend()

titles = ['Died in Class 3', 'Died in Class 1', 'Died in Class 2', 'Survived in Class 3','Survived in Class 1', 'Survived in Class 2']
for ax,title in zip(g.axes.flatten(),titles):
    ax.set_title(title)

plt.show()
```
